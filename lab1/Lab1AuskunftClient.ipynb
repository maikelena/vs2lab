# Telefonauskunft – Client

Dieses Notebook dient als Benutzerschnittstelle für den Auskunftsdienst.

Voraussetzung: Der Server muss bereits in einem zweiten Notebook oder Prozess laufen.

Verfügbare Operationen:
- `get(name)` sucht einen einzelnen Namen
- `getall()` liefert alle Einträge des Telefonbuchs


```python
import clientserver
client = clientserver.Client()
```

# Einzelnen Eintrag abfragen

```python
client.get("Alice")
```

# Unbekannten Namen testen

```python
client.get("Unknown")
```

# Alle Einträge abfragen

```python
client.getall()
```

# Verbindung schließen

Am Ende sollte der Client-Socket sauber geschlossen werden.

```python
client.close()
```
