# Telefonauskunft Client Part
Notebook zum Client der Telefonauskunft.

Operationen:
GET bekommt als Parameter einen Namen und sucht diesen in der Telefon-Datenbank. Das Ergebnis wird über den Socket an den Client Prozess zurück gesendet.
GETALL sendet alle Einträge der Telefon-Datenbank per Socket an den Client Prozess zurück.

Schritt 1: telefonauskunft importieren und Client-Objekt erzeugen

```python
import telefonauskunft
client = telefonauskunft.Client()
```

Schritt 2: Eintrag für Chloe abfragen

```python
client.get("Chloe")
```

Zwischentest: Unbekannten Namen abfragen

```python
client.get("Unknown")
```

Schritt 3: Alle Einträge abfragen

```python
client.getall()
```

Schritt 4: Verbindung schließen

```python
client.close()
```
